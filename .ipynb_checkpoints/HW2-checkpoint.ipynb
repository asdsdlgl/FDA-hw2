# Report and Decision Tree

這份資料主要是探討天氣，主要分成會不會下雨，會為Rain不會則Sun，以下為建立的10000筆特定資料的產生與判斷標準

資料特徵 :

    Temperature : -10 - 40
    WindSpeed   :   0 - 30
    Pa          : 750 - 1030
    Humidity    :   0 - 100

判斷標準 :

    Pa <= 950         && Humidity >= 80
    WindSpeed >= 25   && Humidity >= 60
    Temperature >= 30 && Humidity >= 40
    Pa <= 850
    
其中之一條件達成則Rain，反之為Sun
    

# Data Generation

```python
import pandas as pd
import numpy as np
import random
```

```python
df = pd.DataFrame()
df['Temperature'] = ((np.random.random(10000)*50)-10).astype(int) #temp -10-40
df['WindSpeed'] = ((np.random.random(10000)*30)).astype(int) #wind 0-30
df['Pa'] = ((np.random.random(10000)*280)+750).astype(int) #pa 750-1030
df['Humidity'] = ((np.random.random(10000)*100)).astype(int) #0-100
#conditions = [
#    (df['Temperature'] >= 30) & (df['Humidity'] >= 80),
#    (df['Humidity'] >=60) & (df['WindSpeed'] >= 25),
#    (df['Pa'] <= 850)]
#choices = ['1', '2', '3']
#df['Rain'] = np.select(conditions, choices, default='0').astype(int)
df['Rain'] = np.where(((df['Temperature'] >= 30) & (df['Humidity'] >= 40))|((df['Humidity'] >=60) & (df['WindSpeed'] >= 25))|(df['Pa'] <= 850)|((df['Pa'] <= 950)&(df['Humidity'] >= 80)), 'rain', 'sun')
#df.groupby('Rain').size()
df
```

```python
df.to_csv('weather.csv')
```

# Construct Decision Tree

```python
from sklearn import tree
from sklearn.model_selection import train_test_split
df.head()
Y = df['Rain']
X = df.loc[:, ['Temperature', 'WindSpeed', 'Pa', 'Humidity']]
train_X, test_X, train_y, test_y = train_test_split(X, Y, test_size = 0.3)
deTreeModule = tree.DecisionTreeClassifier(max_depth=4)
deTreeModule.fit(train_X, train_y)
```

```python
from sklearn import metrics
test_y_predicted = deTreeModule.predict(test_X)
accuracy = metrics.accuracy_score(test_y, test_y_predicted)
accuracy
```

```python
import graphviz
import PIL.Image
class_names=['Rain', 'Sun']
data = tree.export_graphviz(deTreeModule, out_file=None,
                feature_names=test_X.columns,  
                class_names=class_names,  
                filled=True, rounded=True,  
                special_characters=True)

graph = graphviz.Source(data)
graph.render('decision_tree', view=True)
graph
```

# Report Discussion

基本上，我是將資料分成7000筆去train，3000筆當test data，其decision tree的結果跟我的判斷標準接近，雖然仍有些許誤差，有上述的準確度accuracy基本上都有95%以上，而其差異主要是像是我定義Pa <= 850 ，而其判斷標準為Pa <= 850.5，這似乎有顯現他電腦判斷上是以浮點數的形式去做判斷的，而我是以整數做處理，這之間可能有誤差，此外，我的tree中有個判斷實際上並沒有完全被寫入，就是Humidity>=80的情況判斷，而這也使得資料的正確性有誤差，舉例來說Humidity=80、Pa=950、windSpeed=Temperature=10的時候，應該為Rain但根據此tree為Sun，也許是trainingData剛好沒判斷到此情形，也有可能是我將tree的depth最大深度訂為4的原因，透過將depth深度加深，應該可以使accuracy的正確度上升，而此樹可能也是在深度4時根據此training data所做出的正確性最高的決策樹了
